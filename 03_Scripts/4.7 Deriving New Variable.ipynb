## Creating Price flags using a user defined function

```python
 # define the function to label products by price range
    
def price_label(row):
    
    if row ['prices'] <= 5:
        return 'Low-range product'
    elif (row ['prices']> 5) and (row['prices']<= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High range'
    else: return 'Not enough data'    
```

```python
# make a column for the flags and run the function on all rows of our dataframe

df['price_range'] = df.apply(price_label, axis= 1)
```

```python
# check the results of the function

df['price_range'].value_counts(dropna= False)
```

```python
# check the maximum price

df['prices'].max()
```

## creating the price flags using the loc function

```python
# first condition

df_merged.loc[df_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# second condition

df_merged.loc[(df_merged['prices'] <= 15) & (df_merged['prices'] > 5), 'price_range_loc'] = 'Mid range product'
```

```python
# third condition

df_merged.loc[df_merged['prices'] <= 5 , 'price_range_loc'] = 'Low range product'
```

```python
# check the results of the function

df_merged['price_range_loc'].value_counts(dropna= False)
```

## Make a column showing the busiest days of the week

```python
# first find the days that are the busiest

df_merged['orders_day_of_week'].value_counts(dropna= False)
```

```python
# make a list called results that will be appended by our for loop, denoting 0- most busy, 4- least, all else -regular

result = []

for value in df_merged['orders_day_of_week']:
    if value == 0:
        result.append('Busiest Day')
    elif value == 4:
        result.append('Least Busy')
    else: 
        result.append('Regularly Busy')
```

```python
# print the result list 

result
```

```python
# make a new column in the df that contains the results list

df_merged['busiest_day'] = result
```

```python
# check the values

df_merged['busiest_day'].value_counts(dropna= False)
```

## Step 2:

Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

### The two most busy days are 0 and 1, the two least busy days are 3 and 4

```python
# Step 1: Define lists of most and least busy days so that more days could be added or removed as needed without a new loop

most_busy = [0,1]
least_busy = [3,4]
```

```python
# Step 2: make a new list that is appended by the for loop, adding a corresponding tags to the values in our lists

results_b = []

for value in df_merged['orders_day_of_week']:
    if value in most_busy:
        results_b.append('Busiest days')
    elif value in least_busy:
        results_b.append('Slowest days')
    else:
        results_b.append('Regularly Busy')
```

```python
# Step 3: Set the results_b list to a new column in our dataframe

df_merged['busiest_days'] = results_b
```

## Step 4

When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# check the head to find the column which contains this information

df_merged.head()
```

```python
# the column we need is order_hour_of_day, check value to find the busiest hours

df_merged['order_hour_of_day'].value_counts(dropna= False)
```

### This can be achieved in several ways, but I will use the loc function for this task

```python
# for clarity later on, I will first delcare lists with my hours in categories 

most_orders= [10,11,14,15,13,12,16,9]
average_orders =[17,8,18,19,20,7,21,22]
least_orders = [23, *range(7)] 
```

### Now I will create my conditions to add flags to a new column called 'busiest_period_of_day'

```python
# first condition busiest hours

df_merged.loc[df_merged['order_hour_of_day'].isin(most_orders), 'busiest_period_of_day'] = 'Most Orders'
```

```python
# second condition average hours

df_merged.loc[df_merged['order_hour_of_day'].isin(average_orders), 'busiest_period_of_day'] = 'Average Orders'
```

```python
# third condition least hours

df_merged.loc[df_merged['order_hour_of_day'].isin(least_orders), 'busiest_period_of_day'] = 'Fewest Orders'
```
